### Summary & Conclusion

1. **No Data from New York**: The dataset does not include any accident data from New York.
2. **City Accident Frequency**: Less than 8% of the cities report more than 1,000 yearly accidents.
3. **Low Incident Cities**: Approximately 1,023 cities have reported only 1 accident.
4. **Peak Accident Times**: The majority of accidents occur between 6 AM - 10 AM and 3 PM - 6 PM.
5. **Weekend Trends**: Accident rates are lower on weekends.
6. **Sunday Peak Hours**: On Sundays, the peak accident times are between 11 AM - 4 PM.
7. **Missing Data**: Data for the year 2023 is missing.
8. **Data Quality Issues**: There are issues with data from Source2 and Source3, indicating potential inconsistencies or errors.
